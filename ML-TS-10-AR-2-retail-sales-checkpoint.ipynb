# STEP 1- Quick Check for Autocorrelation

- ** method 1 - using pandas lag_plot **

```python
lag_plot(sales_data['sales']);
```

Because we don’t have many data points, this particular lag_plot() doesn’t look terribly convincing, but there is some correlation in there (along with some possible outliers).

- ** Method 2 - using corr() **

```python
values    = pd.DataFrame(sales_data.values)
dataframe = pd.concat([values.shift(1), values], axis=1)

dataframe.columns = ['t-1', 't+1']

result = dataframe.corr()
print(result)
```

This is a good confirmation for the plot above.

It shows a somewhat strong positive correlation (0.54) between the observation and the lag=1 value.

- ** Method 3 - using autocorrelation_plot **

```python
pd.plotting.autocorrelation_plot(sales_data['sales']);
```

The dark horizontal line at zero just denotes the zero line, the lighter full horizontal lines is the 95% confidence level and the dashed horizontal lines are 99% confidence levels, which means that correlations are more significant if they occur at those levels.

there’s some significant correlation between t=1 and t=12 (roughly) with significant decline in correlation after that timeframe.  Since we are looking at monthly sales data, this seems to make sense with correlations falling off at the start of the new fiscal year.

# STEP 2- Check for stationarity

```python
decomposed = seasonal_decompose(sales_data['sales'], model='additive')
x = decomposed.plot() 
```

we have a time series that has a trend (2nd panel from top) and has seasonality (third panel from top).  

Let’s make it stationary by removing/reducing trend and seasonality.

** use differencing **

```python
sales_data['stationary']=sales_data['sales'].diff()
```

```python
plt.plot(sales_data['stationary']);
```

** seasonal_decompose() on this new data gives us: **

```python
sales_data = sales_data.dropna()
```

```python
decomposed = seasonal_decompose(sales_data['stationary'], model='additive')
x = decomposed.plot() #See note below about this
```

there’s still some trend and even some seasonality,

# STEP 3 - Forecasting 

```python
# create train/test datasets
X = sales_data['stationary'].dropna()
```

```python
# Extract train and test data
train_data = X[1:len(X) - 12]
test_data  = X[X[len(X) - 12:]]
 
# train the autoregression model
model = AR(train_data)
model_fitted = model.fit()
```

```python
print('The lag value chose is: %s' % model_fitted.k_ar)
 
print('The coefficients of the model are:\n %s' % model_fitted.params)
```

```python
# make predictions 
predictions = model_fitted.predict(
    start=len(train_data), 
    end=len(train_data) + len(test_data)-1, 
    dynamic=False)
 
# create a comparison dataframe
compare_df = pd.concat(
    [sales_data['stationary'].tail(12),
    predictions], axis=1).rename(
    columns={'stationary': 'actual', 0:'predicted'})
 
#plot the two values
compare_df.plot()
```

```python
r2_score(sales_data['stationary'].tail(12), predictions)
```
