### Identifying Trends in Time Series

```python
# all column pandas Dataframe
# diet = df['diet']
# gym  = df['gym']
# fin  = df['finance']

# all column pandas Series
diet = df['diet']
gym  = df['gym']
fin  = df['finance']

plt.figure(figsize=(20,10))

diet.rolling(12).mean().plot(linewidth=5, fontsize=20, color='k', label='diet')
gym.rolling(12).mean().plot(linewidth=5, fontsize=20, color='r', label='gym')
fin.rolling(12).mean().plot(linewidth=5, fontsize=20, color='y', label='finance')

plt.legend(loc='best')
plt.xlabel('Year', fontsize=20);
```

- we have removed most of the seasonality compared to the previous plot.
- diet potentially has some form of seasonality, whereas gym is actually increasing!
